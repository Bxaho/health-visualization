# Health Visualization 

```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python

# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("./Health_heart_experimental.csv")
df.head(10)
```

# LINE GRAPH

This shows that the graph is dispersed, indicating that there is no linear relationship and that age has no effect on SysBP; rather, it is dependent on an individual's health.

```python
age = [21,30,35,39,61]
bmi = [21.0,21.0,20.0,33.0,32.0]

# create a line chart, age on x-axis, bmi on y-axis
plt.plot(age, bmi, color='green', marker='+', linestyle='dotted')

# add a title
plt.title("HEART HEALTH ")

# add a label to the y-axisage
plt.ylabel("Bmi")
plt.xlabel("age")
plt.show()
```

# BAR PLOT


Bar plot for SysBP of first 10 persons

```python
persons = [f"Per {i+1}" for i in range(10)]
SysBP = [141, 109, 112, 106, 140, 115, 125, 111, 144, 113]

# Create the bar plot
plt.bar(range(len(persons)), SysBP)

# Add title and labels
plt.title("SysBP of 10 People")
plt.ylabel("SysBP")
plt.xlabel("Persons")

plt.xticks(range(len(persons)), persons)

# Display the plot
plt.show()
```

# PIE CHART

This pie chart shows that 52.68% of the people need no treatment they are healthy but 47.32% need treatment it says they are not healthy

```python
df['indication'].value_counts().plot(kind='pie',autopct='%0.2f')
plt.title("Indication")
```

# BOX PLOT

According to this statistics, the heart rate's meadian is approximately 140.According to this, the average person's heart rate is between 105 and 170 beats per minute.NO outliers, or individuals with abnormal heart rates, are present.

```python
sns.boxplot(df['HR'])
```

# HISTORGRAM

The age distribution of those with and without health issues is depicted in the graph. Compared to those who may have health difficulties, people without health problems are often older and range in age.



```python
sns.kdeplot(df[df['indication'] == 0]['age'], label='No Indication')
sns.kdeplot(df[df['indication'] == 1]['age'], label='Indication')

plt.title('Age Distribution by Indication')
plt.xlabel('Age')
plt.ylabel('Density')

plt.legend()

plt.show()
```

# heat map

The heat map between sex and indication is displayed here, with 0 denoting male and 1 denoting female. For indication, 0 denotes no health difficulties and 1 denotes health issues. Based on the data, it is evident that males are more likely than females to have health issues.

```python
sns.heatmap(pd.crosstab(df['sex'],df['indication']))
```
